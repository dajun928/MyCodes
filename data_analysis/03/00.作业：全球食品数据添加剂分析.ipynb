```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "FoodFacts.csv"
```

```python
# 1. 读取数据集
df_obj = pd.read_csv(filename, usecols=['countries_en', 'additives_n'])
df_obj
```

```python
# 2. 数据清洗
df_obj.dropna(inplace=True)
print(df_obj.head(10))
```

```python
# 从DataFrame堆里获取 Series的行和列数据
# 取所有国家的数据
country_arr = df_obj['countries_en'].values
# 取所有添加剂的数据
add_arr = df_obj["additives_n"].values
```

```python
# 将国家名做为行索引，添加剂做为值，创建Series对象
ser_obj = pd.Series(add_arr, index= country_arr)
print(ser_obj.head(10))
```

```python
# 将国家名去重，防止重复迭代
country_set = set(country_arr)
print(country_set)
```

```python
ser_obj['France'].sum()
```

```python
# 迭代每个国家，取出该国家在Series对象里的添加剂总和，并返回字典 每个国家对应添加剂的总和
dict_obj = {country : ser_obj[country].sum() for country in country_set}
```

```python
dict_obj
```

```python
# 再使用字典创建Series对象，并排序取前10个数据
data = pd.Series(dict_obj).sort_values(ascending=False)[:10]
print(data)
```

```python

data.plot.bar()
plt.show()
```

```python
#pd.read_csv()
data.index.name = "country"
data.to_csv("Footfaces_top10.csv")
```
