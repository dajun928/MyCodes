```python
# 

# 1
bins = [-np.inf, -7, -5, -3, 0, 3, 5, 7, np.inf]
bi = pd.cut(data['p_change'], bins)
bi
# 2
```

```python
# 

# 1
week = pd.to_datetime(data.index).weekday
```

```python
data
```

```python
# 2
data['week'] = week
```

```python
# 3
data['positive'] = np.where(data['p_change'] > 0, 1, 0)
```

```python
data
```

```python
# 
cross_tab = pd.crosstab(data['week'], data['positive'])
```

```python
cross_tab
```

```python
cross_tab.sum(1).astype(float)
```

```python
tab = cross_tab.div(cross_tab.sum(1).astype(float), axis=0)
```

```python
tab
```

```python
tab.plot(kind='bar', stacked=True)
```
