```python
import numpy as np
```

```python
filename = "国际数据主要国家(地区)年度数据.csv"
```

```python
arr = np.genfromtxt(filename, delimiter=",", dtype=bytes)
print(arr)
```

```python
# 1. 取出所有国家的所有年份的人口数据
people_arr = arr[5:-2, 1:]
# 数据清洗，将空字符串替换为 b"0"，再修改数据类型为int
people_arr = np.where(people_arr == b"", b"0", people_arr).astype(np.uint32)
print(people_arr)
```

```python
# 2. 获取所有国家的名称
country_list = [country.decode("utf-8") for country in arr[4, 1:]]
#print(country_list)
country_arr = np.array(country_list)
print(country_arr)
```

```python
# 3. 获取年份的名称
year_list = [year.decode("utf-8") for year in arr[5:-2, 0]]
year_arr = np.array(year_list)
print(year_arr)
```

## 1. 统计2015年各个国家人口数据

```python
year = "2015年"
#print(year_list.index(year))
# 查找数组中指定元素的下标，返回一个二维数组，通过[0, 0]
# 表示people_arr的行
year_index = np.argwhere(year_arr == year)[0, 0]

for country, people in zip(country_arr, people_arr[year_index]):
    print("{} : {}".format(country, people))
```

## 2. 统计朝鲜历史各个时期人口数据

```python
country = "朝鲜"
country_index = np.argwhere(country_arr == country)[0, 0]
print(country_index)
```

```python
# print(people_arr[:, country_index])

print("{}人口数据".format(country))
for year, people in zip(year_arr, people_arr[:, country_index]):
    print("{} : {}".format(year, people))
```

## 3. 统计缅甸2014年的人口数据

```python
year = "2014年"
country = "缅甸"

year_index = np.argwhere(year_arr == year)[0, 0]
country_index = np.argwhere(country_arr == country)[0, 0]
```

```python
print("{} 在 {}的人口数据：{}".format(country, year, people_arr[year_index, country_index]))
```

## 4. 统计每一个国家历史平均人口数据

```python
for country, people in zip(country_arr, np.mean(people_arr, axis=0)):
    print("{:10}\t历史平均人口数据：\t {}".format(country, people))
```

## 5. 统计亚洲2015年总人口，及平均人口

```python
year = "2015年"
year_index = np.argwhere(year_arr == year)[0, 0]

print("亚洲在 {} 的总人口: {:.2f}亿 和 平均人口 :{:.2f} ".format(year, np.sum(people_arr[year_index]) / 10 ** 8, np.mean(people_arr[year_index])))
```

## 6. 统计印度、柬埔寨、阿富汗在2011、2012、2013年总人口及平均人口

```python
country_list = ["印度", "柬埔寨", "阿富汗"]
year_list = ["2011年", "2012年", "2013年"]
```

```python
country_index_list = [np.argwhere(country_arr == country)[0, 0] for country in country_list]
year_index_list = [np.argwhere(year_arr == year)[0, 0] for year in year_list]
print(country_index_list)
print(year_index_list)
```

```python
# 取不同国家在三年的数据
## 先获取所有国家的 year_list 每一年数据 arr[[0, 2, 4]]
year_list_people_arr = people_arr[year_index_list]
#print(year_list_people_arr)

for country, country_index in zip(country_list, country_index_list):
    print("{} 在 {}".format(country, ",".join(year_list)), end=' ')
    print("总人口 : {}".format(np.sum(year_list_people_arr[:, country_index])), end=' ')
    print("平均人口: {:.2f}".format(np.mean(year_list_people_arr[:, country_index])))
```

```python
# 取不同年份在三个国家的数据

## 先获取所有国家的 year_list 每一年数据 arr[[0, 2, 4]]
country_list_people_arr = people_arr[:, country_index_list]
#print(country_list_people_arr)

for year, year_index in zip(year_list, year_index_list):
    print("{} 在 {}".format(year, ",".join(country_list)), end=' ')
    print("总人口 : {}".format(np.sum(country_list_people_arr[year_index])), end=' ')
    print("平均人口: {:.2f}".format(np.mean(country_list_people_arr[year_index])))
```

## 7. 统计任意两个国家之间的人口差数据

```python
country1 = "阿富汗"
country2 = "以色列"
year = "2008年"

country1_index = np.argwhere(country_arr == country1)[0, 0]
country2_index = np.argwhere(country_arr == country2)[0, 0]

people_2008_arr = people_arr[np.argwhere(year_arr == year)[0, 0]]
print(people_2008_arr)
```

```python
print("{} 的 {} 的人口数据: {} ".format(year, country1, people_2008_arr[country1_index]))
print("{} 的 {} 的人口数据: {} ".format(year, country2, people_2008_arr[country2_index]))
print("两个国家的差值为: {}".format(people_2008_arr[country1_index] - people_2008_arr[country2_index]))
```

## 8 .统计2012年亚洲人口数量排名前10的国家

```python
year = "2012年"
people_2012_arr = people_arr[np.argwhere(year_arr == year)[0, 0]]
print(people_2012_arr)
```

```python
# Python的写法

people_dict = dict(enumerate(people_2012_arr))

# sorted() 对字典排序：key表示排序的依据，传入items()每个元素，并按值进行怕需
sorted_data = sorted(people_dict.items(), key = lambda n : n[1])
print(sorted_data)

sorted_index_arr = np.array(sorted_data)[:, 0]
print(sorted_index_arr)
```

```python
# ndarray的写法

# 对数组排序，并返回排序后的索引
sorted_index_arr = people_2012_arr.argsort()
print(sorted_index_arr)
```

```python
# 根据索引取值
sorted_country = country_arr[sorted_index_arr]
sorted_people = people_2012_arr[sorted_index_arr]
```

```python
data = list(zip(sorted_country, sorted_people))[-10:][::-1]


for country, people in data:
    print(country, people)
```
