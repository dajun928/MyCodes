### 散点图的绘制 scatter
- 适合观察变量的分布
- 适合观察变量之间的关系

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['Arial Unicode MS']
import random
```

```python
x = range(60)
y = [i*2 + random.random() for i in x] 
y1 = [random.uniform(5,9) for i in x] # 均匀分布
```

```python
plt.figure(figsize=(10,8), dpi = 200)

plt.subplot(121)

plt.scatter(x,y)

plt.title('观察变量关系')

plt.subplot(122)

plt.scatter(x,y1)

plt.title('观察分布')
plt.show()
```

### k线图

```python
import pandas as pd
data = pd.read_hdf("../../data/stock_plot/day_open.h5")[:100]
data1 = pd.read_hdf("../../data/stock_plot/day_close.h5")[:100]
data2 = pd.read_hdf("../../data/stock_plot/day_high.h5")[:100]
data3 = pd.read_hdf("../../data/stock_plot/day_low.h5")[:100]

day = pd.concat([data["000001.SZ"], data1["000001.SZ"], data2["000001.SZ"],
data3["000001.SZ"]], axis=1)

day.columns = ["open", "close", "high", "low"]
day = day.reset_index().values
```

```python
day
```

```python
from matplotlib.finance import candlestick_ochl

# plt.figure(figsize=(10,8), dpi = 100)

# axe = plt.subplot(111)
fig, axe = plt.subplots(nrows=1, ncols=1, figsize=(10,8), dpi = 100)

candlestick_ochl(axe, day, width=0.5,colorup='r', colordown='g', alpha=0.5)
plt.show()
```
