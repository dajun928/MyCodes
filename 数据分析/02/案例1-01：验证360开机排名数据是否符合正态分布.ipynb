### 首先读入 用户开机数据

```python
filename = './startTime.csv'
data = np.genfromtxt(filename, delimiter=",")
```

### 绘制直方图判断是否符合 正态分布

```python
plt.figure(figsize=(10,5), dpi=100)
plt.title('开机数据分布情况')
x = data
x.sort()
x = np.unique(x)
plt.xticks(x[::2])
plt.hist(data, bins=x[::2])
plt.grid(alpha=0.5,linestyle='--')
plt.show()
```
