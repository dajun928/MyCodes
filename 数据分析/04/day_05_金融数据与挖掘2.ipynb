```python
stock_day = pd.read_csv("./data/stock_day/stock_day.csv")
```

```python
stock_day = stock_day.sort_index()
```

```python
# 先将K线图画出
ochl = stock_day[['open', 'close', 'high', 'low']]
```

```python
ochl.index = [i for i in range(643)]
```

```python
# 重设索引，是的索引变成一列数据
arr = ochl.reset_index().values[:200, :]
print(arr)
```

```python
fig, axes = plt.subplots(figsize=(20,8), dpi=100)
print(axes)
```

```python
candlestick_ochl(axes, arr, colordown='g', colorup='r', width=0.5)
```
