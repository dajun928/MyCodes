```python
# 500 * 504
stock_day_rise = np.random.normal(0, 1, (500, 504))
```

```python
# 把行索引编程 股票0 股票1 。。。。。
# index:可以指定行索引的名字
# 构建名字的列表 (500, 504)
indexs = ['股票' + str(i) for i in range(stock_day_rise.shape[0])]

pd.DataFrame(stock_day_rise, index=indexs)
```

```python
# 查看股票涨跌浮动是不是符合正态分布
data['p_change'].hist(bins=80)
```

```python
plt.show()
```

```python
# 分组

st = pd.qcut(np.abs(data['p_change']), 10)
```

```python
# 统计分组次数
st.value_counts()
```

```python
data
```

```python
# 生成哑变量矩阵

# 1、先分组
bins = [-np.inf, -7, -5, -3, 0, 3, 5, 7, np.inf]
bi = pd.cut(data['p_change'], bins)
bi
# 2、生成
```

```python
# 将分组之后的结果放入
dummary = pd.get_dummies(bi, prefix="split_rise")
```

```python
# 目的：把数据离散化统计一下
pd.concat([data, dummary], axis=1)
```

```python
# 判断股票的涨跌与什么有关系

# 1、生成星期的列数据
week = pd.to_datetime(data.index).weekday
```

```python
data
```

```python
# 2、增加星期几数据
data['week'] = week
```

```python
# 3、标记涨跌幅的好坏
data['positive'] = np.where(data['p_change'] > 0, 1, 0)
```

```python
data
```

```python
# 交叉表
cross_tab = pd.crosstab(data['week'], data['positive'])
```

```python
cross_tab
```

```python
cross_tab.sum(1).astype(float)
```

```python
tab = cross_tab.div(cross_tab.sum(1).astype(float), axis=0)
```

```python
tab
```

```python
tab.plot(kind='bar', stacked=True)
```

```python
plt.show()
```
