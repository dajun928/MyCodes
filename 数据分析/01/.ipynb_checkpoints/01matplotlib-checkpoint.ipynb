```python
import matplotlib.pyplot as plt
import random
```

```python
# 
# plt.figure(figsize=(20, 8), dpi=100)
# axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 
x = list(range(0, 60))
y = [random.uniform(15, 18) for i in range(0, 60)]

# 
y_beijing = [random.uniform(1, 3) for i in range(0, 60)]

# 
# 
axes[0].plot(x, y, label="")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="")

x_name = ["11{}".format(i) for i in range(0, 60)]

# show
# 
plt.xticks(x[::5], x_name[::5])
plt.yticks(list(range(0, 45))[::5])

# 
# 
plt.legend(loc="best")
plt.xlabel("")
plt.ylabel("")
plt.title("1101159")

plt.show()
```

```python
# 
plt.figure(figsize=(20, 8), dpi=100)

#  
movie_name = ['3','','','','','','','','','','']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# x
x = range(len(movie_name))

# bar
plt.bar(x, y, width=0.2, color=['b','r','g','y','c','m','y','k','c','g','g'])

#X
plt.xticks(x, movie_name)

plt.show()
```

```python
# 
plt.figure(figsize=(20, 8), dpi=100)

# 
movie_name = ['3','','']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x = range(len(movie_name))


# 
plt.bar(x, first_day, width=0.2, label="")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, label="")

plt.legend(loc="best")

# 
plt.xticks([i+0.1 for i in x], movie_name)
plt.title("")

# 
plt.show()
```
