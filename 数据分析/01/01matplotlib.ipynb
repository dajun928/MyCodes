```python
# 某城市一小时内的温度变化情况
# plt.figure(figsize=(20, 8), dpi=100)
# axes列表
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 准备温度数据
x = list(range(0, 60))
y = [random.uniform(15, 18) for i in range(0, 60)]

# 准备北京的温度
y_beijing = [random.uniform(1, 3) for i in range(0, 60)]

# 画折线图
# 添加线条注释，还需要增加显示
axes[0].plot(x, y, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="北京")

x_name = ["11点{}分".format(i) for i in range(0, 60)]

# 图的名字，标签，美化过程，在show之前使用
# 美化刻度值时候，必须使用数值类型的数字
plt.xticks(x[::5], x_name[::5])
plt.yticks(list(range(0, 45))[::5])

# 美化标题，横坐标描述，纵坐标描述
# 增加折线注释
plt.legend(loc="best")
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("某周11点0分到11点59分的不同地区温度变化情况")

plt.show()
```

```python
# 画条形图显示电影票房对比
plt.figure(figsize=(20, 8), dpi=100)

# 准备好数据， 电影名称，电影票房
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# 显示x坐标必须是数字
x = range(len(movie_name))

# 通过bar去显示条形图
plt.bar(x, y, width=0.2, color=['b','r','g','y','c','m','y','k','c','g','g'])

#美化X坐标显示
plt.xticks(x, movie_name)

plt.show()
```

```python
# 画条形图显示电影票房对比
plt.figure(figsize=(20, 8), dpi=100)

# 准备数据
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x = range(len(movie_name))


# 显示三部电影第一天的票房
plt.bar(x, first_day, width=0.2, label="首日票房")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, label="首周票房")

plt.legend(loc="best")

# 修改刻度的位置和值
plt.xticks([i+0.1 for i in x], movie_name)
plt.title("每日每周票房对比")

# 显示
plt.show()
```
